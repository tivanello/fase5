```python
# ============================================================
# BLOCO P10.15 — VALIDAR PEDRA 23 x PEDRA 2023 POR RA
# Objetivo:
# - consolidar por RA
# - trazer Pedra 23 e Pedra 2023 lado a lado
# - classificar a situação entre as duas colunas
# ============================================================

df_tmp = df_fase5_p10[["RA", "Pedra 23", "Pedra 2023"]].copy()

# padroniza texto
for c in ["Pedra 23", "Pedra 2023"]:
    df_tmp[c] = df_tmp[c].astype("string").str.strip()

# função para pegar o primeiro valor não nulo
def primeiro_nao_nulo(serie):
    serie = serie.dropna()
    return serie.iloc[0] if len(serie) > 0 else pd.NA

# consolida por RA
df_validacao_2023 = (
    df_tmp
    .groupby("RA", as_index=False)
    .agg({
        "Pedra 23": primeiro_nao_nulo,
        "Pedra 2023": primeiro_nao_nulo
    })
)

# coluna de validação
def validar_pedra_2023(row):
    p23 = row["Pedra 23"]
    p2023 = row["Pedra 2023"]

    if pd.isna(p23) and pd.isna(p2023):
        return "AMBAS_VAZIAS"
    elif pd.notna(p23) and pd.notna(p2023) and p23 == p2023:
        return "IGUAL"
    elif pd.notna(p23) and pd.isna(p2023):
        return "COMPLEMENTAR_SO_PEDRA23"
    elif pd.isna(p23) and pd.notna(p2023):
        return "COMPLEMENTAR_SO_PEDRA2023"
    else:
        return "CONFLITO"

df_validacao_2023["validacao"] = df_validacao_2023.apply(validar_pedra_2023, axis=1)

print("Resumo da validação por RA:")
display(
    df_validacao_2023["validacao"]
    .value_counts(dropna=False)
    .rename_axis("validacao")
    .reset_index(name="quantidade")
)

print("\nAmostra da base consolidada por RA:")
display(df_validacao_2023.head(30))
```

```python
# ============================================================
# BLOCO P10.15 — VALIDAR PEDRA 23 x PEDRA 2023 POR RA
# Objetivo:
# - consolidar por RA
# - trazer Pedra 23 e Pedra 2023 lado a lado
# - classificar a situação entre as duas colunas
# ============================================================

df_tmp = df_fase5_p10[["RA", "Pedra 23", "Pedra 2023"]].copy()

# padroniza texto
for c in ["Pedra 23", "Pedra 2023"]:
    df_tmp[c] = df_tmp[c].astype("string").str.strip()

# função para pegar o primeiro valor não nulo
def primeiro_nao_nulo(serie):
    serie = serie.dropna()
    return serie.iloc[0] if len(serie) > 0 else pd.NA

# consolida por RA
df_validacao_2023 = (
    df_tmp
    .groupby("RA", as_index=False)
    .agg({
        "Pedra 23": primeiro_nao_nulo,
        "Pedra 2023": primeiro_nao_nulo
    })
)

# coluna de validação
def validar_pedra_2023(row):
    p23 = row["Pedra 23"]
    p2023 = row["Pedra 2023"]

    if pd.isna(p23) and pd.isna(p2023):
        return "AMBAS_VAZIAS"
    elif pd.notna(p23) and pd.notna(p2023) and p23 == p2023:
        return "IGUAL"
    elif pd.notna(p23) and pd.isna(p2023):
        return "COMPLEMENTAR_SO_PEDRA23"
    elif pd.isna(p23) and pd.notna(p2023):
        return "COMPLEMENTAR_SO_PEDRA2023"
    else:
        return "CONFLITO"

df_validacao_2023["validacao"] = df_validacao_2023.apply(validar_pedra_2023, axis=1)

print("Resumo da validação por RA:")
display(
    df_validacao_2023["validacao"]
    .value_counts(dropna=False)
    .rename_axis("validacao")
    .reset_index(name="quantidade")
)
```

```python
# ============================================================
# BLOCO — TRANSFORMAR AS PEDRAS EM BASE LONGA POR ALUNO
# Objetivo:
# - reorganizar as colunas de pedra em formato longitudinal
# - manter apenas registros com pedra preenchida
# ============================================================

df_base = df_fase5_p10.copy()

colunas_pedra = {
    "Pedra 21": 1,
    "Pedra 22": 2,
    "Pedra 23": 3,
    "Pedra 2024": 4,
    "PEDRA_final": 5
}

bases = []

for col, ordem in colunas_pedra.items():
    df_temp = df_base[["RA", col]].copy()
    df_temp = df_temp.rename(columns={col: "pedra"})
    df_temp["pedra_referencia"] = col
    df_temp["ordem_referencia"] = ordem
    bases.append(df_temp)

df_pedra_long = pd.concat(bases, ignore_index=True)

df_pedra_long = df_pedra_long[["RA", "pedra_referencia", "ordem_referencia", "pedra"]].copy()

# remove linhas sem pedra
df_pedra_long = df_pedra_long[df_pedra_long["pedra"].notna()].copy()

print("Dimensão da base longa:")
print(df_pedra_long.shape)

print("\nAmostra:")
display(df_pedra_long.head(20))

print("\nNulos por coluna:")
print(df_pedra_long.isna().sum())
```

```python
# ============================================================
# BLOCO — LISTAR AS PEDRAS EXISTENTES
# Objetivo:
# - mostrar quais valores existem na coluna de pedra
# - incluir a quantidade de ocorrências de cada uma
# ============================================================

print("Pedras existentes:")
display(df_pedra_long["pedra"].value_counts(dropna=False).rename_axis("pedra").reset_index(name="quantidade"))
```

```python
# ============================================================
# BLOCO — PADRONIZAR AS PEDRAS DA BASE LONGA E CRIAR SCORE
# Objetivo:
# - corrigir grafias diferentes da mesma pedra
# - remover valores inválidos como INCLUIR
# - criar score ordinal para análise de evolução
# ============================================================

df_pedra_long = df_pedra_long.copy()

df_pedra_long["pedra"] = df_pedra_long["pedra"].astype("string").str.strip()

df_pedra_long["pedra"] = df_pedra_long["pedra"].replace({
    "Agata": "Ágata",
    "AGATA": "Ágata",
    "Topazio": "Topázio",
    "TOPAZIO": "Topázio",
    "INCLUIR": pd.NA
})

df_pedra_long = df_pedra_long[df_pedra_long["pedra"].notna()].copy()

mapa_score_pedra = {
    "Quartzo": 1,
    "Ágata": 2,
    "Ametista": 3,
    "Topázio": 4
}

df_pedra_long["score_pedra"] = df_pedra_long["pedra"].map(mapa_score_pedra)

print("Pedras existentes após padronização:")
display(
    df_pedra_long["pedra"]
    .value_counts(dropna=False)
    .rename_axis("pedra")
    .reset_index(name="quantidade")
)

print("\nAmostra com score_pedra:")
display(df_pedra_long.head(20))

print("\nNulos em score_pedra:")
print(df_pedra_long["score_pedra"].isna().sum())
```

```python
# ============================================================
# BLOCO — GERAR O SCORE FINAL DA PEDRA FINAL POR ALUNO
# Objetivo:
# - pegar a linha da PEDRA_final
# - gerar o score final de cada aluno
# ============================================================

df_score_final = (
    df_pedra_long[df_pedra_long["pedra_referencia"] == "PEDRA_final"]
    [["RA", "pedra", "score_pedra"]]
    .copy()
    .rename(columns={
        "pedra": "pedra_final",
        "score_pedra": "score_pedra_final"
    })
    .sort_values("RA")
    .reset_index(drop=True)
)

print("Dimensão da base de score final:")
print(df_score_final.shape)

print("\nAmostra:")
display(df_score_final.head(20))

print("\nDistribuição do score final:")
display(
    df_score_final["score_pedra_final"]
    .value_counts(dropna=False)
    .sort_index()
    .rename_axis("score_pedra_final")
    .reset_index(name="quantidade")
)
```

```python
# ============================================================
# BLOCO — GERAR SCORE INICIAL, FINAL E VARIAÇÃO POR ALUNO
# Objetivo:
# - identificar a primeira e a última pedra de cada aluno
# - calcular a variação final da trajetória
# ============================================================

df_traj = df_pedra_long.copy()

df_traj = df_traj.sort_values(["RA", "ordem_referencia"])

df_inicio_fim = (
    df_traj.groupby("RA")
    .agg(
        pedra_inicial=("pedra", "first"),
        score_inicial=("score_pedra", "first"),
        pedra_final=("pedra", "last"),
        score_final=("score_pedra", "last")
    )
    .reset_index()
)

df_inicio_fim["variacao_score"] = df_inicio_fim["score_final"] - df_inicio_fim["score_inicial"]

print("Amostra da trajetória inicial e final:")
display(df_inicio_fim.head(20))

print("\nDistribuição da variação do score:")
display(
    df_inicio_fim["variacao_score"]
    .value_counts(dropna=False)
    .sort_index()
    .rename_axis("variacao_score")
    .reset_index(name="quantidade")
)
```

```python
# ============================================================
# BLOCO — TRANSFORMAR AS PEDRAS EM BASE LONGA (2022 A 2024)
# Objetivo:
# - reorganizar as colunas de pedra em formato longitudinal
# - considerar apenas o período oficial do Datathon: 2022 a 2024
# - manter apenas registros com pedra preenchida
# ============================================================

df_base = df_fase5_p10.copy()

colunas_pedra = {
    "Pedra 22": 1,
    "Pedra 23": 2,
    "Pedra 2024": 3
}

bases = []

for col, ordem in colunas_pedra.items():
    df_temp = df_base[["RA", col]].copy()
    df_temp = df_temp.rename(columns={col: "pedra"})
    df_temp["pedra_referencia"] = col
    df_temp["ordem_referencia"] = ordem
    bases.append(df_temp)

df_pedra_long = pd.concat(bases, ignore_index=True)

df_pedra_long = df_pedra_long[["RA", "pedra_referencia", "ordem_referencia", "pedra"]].copy()

# remove linhas sem pedra
df_pedra_long = df_pedra_long[df_pedra_long["pedra"].notna()].copy()

print("Dimensão da base longa:")
print(df_pedra_long.shape)

print("\nAmostra:")
display(df_pedra_long.head(20))

print("\nNulos por coluna:")
print(df_pedra_long.isna().sum())
```

```python
# ============================================================
# BLOCO — BASE LONGA DE PEDRAS (2022 A 2024) COM PADRONIZAÇÃO E SCORE
# Objetivo:
# - considerar apenas 2022, 2023 e 2024
# - padronizar nomes das pedras
# - remover valores inválidos
# - criar score ordinal para análise de evolução
# ============================================================

df_base = df_fase5_p10.copy()

colunas_pedra = {
    "Pedra 22": 1,
    "Pedra 23": 2,
    "Pedra 2024": 3
}

bases = []

for col, ordem in colunas_pedra.items():
    df_temp = df_base[["RA", col]].copy()
    df_temp = df_temp.rename(columns={col: "pedra"})
    df_temp["pedra_referencia"] = col
    df_temp["ordem_referencia"] = ordem
    bases.append(df_temp)

df_pedra_long = pd.concat(bases, ignore_index=True)

df_pedra_long = df_pedra_long[["RA", "pedra_referencia", "ordem_referencia", "pedra"]].copy()

df_pedra_long["pedra"] = df_pedra_long["pedra"].astype("string").str.strip()

df_pedra_long["pedra"] = df_pedra_long["pedra"].replace({
    "Agata": "Ágata",
    "AGATA": "Ágata",
    "Topazio": "Topázio",
    "TOPAZIO": "Topázio",
    "INCLUIR": pd.NA
})

df_pedra_long = df_pedra_long[df_pedra_long["pedra"].notna()].copy()

mapa_score_pedra = {
    "Quartzo": 1,
    "Ágata": 2,
    "Ametista": 3,
    "Topázio": 4
}

df_pedra_long["score_pedra"] = df_pedra_long["pedra"].map(mapa_score_pedra)

print("Dimensão da base longa:")
print(df_pedra_long.shape)

print("\nAmostra:")
display(df_pedra_long.head(20))

print("\nDistribuição das pedras:")
display(
    df_pedra_long["pedra"]
    .value_counts(dropna=False)
    .rename_axis("pedra")
    .reset_index(name="quantidade")
)

print("\nNulos em score_pedra:")
print(df_pedra_long["score_pedra"].isna().sum())
```

```python
# ============================================================
# BLOCO — CONSULTA DAS PEDRAS POR RA COM ANÁLISE DOS SCORES
# Objetivo:
# - montar uma visão por aluno com as pedras de 2022, 2023 e 2024
# - converter as pedras em scores ordinais
# - classificar a trajetória do aluno no período
# ============================================================

df_consulta_pedras = df_fase5_p10.copy()

colunas = ["RA", "Pedra 22", "Pedra 23", "Pedra 2024"]
df_consulta_pedras = df_consulta_pedras[colunas].copy()

# padroniza texto
for c in ["Pedra 22", "Pedra 23", "Pedra 2024"]:
    df_consulta_pedras[c] = df_consulta_pedras[c].astype("string").str.strip()

# padroniza grafias e remove valor inválido
mapa_padrao = {
    "Agata": "Ágata",
    "AGATA": "Ágata",
    "Topazio": "Topázio",
    "TOPAZIO": "Topázio",
    "INCLUIR": pd.NA
}

for c in ["Pedra 22", "Pedra 23", "Pedra 2024"]:
    df_consulta_pedras[c] = df_consulta_pedras[c].replace(mapa_padrao)

# função para pegar o primeiro valor não nulo por RA
def primeiro_nao_nulo(s):
    s = s.dropna()
    return s.iloc[0] if len(s) > 0 else pd.NA

# consolida por RA
df_consulta_pedras = (
    df_consulta_pedras
    .groupby("RA", as_index=False)
    .agg({
        "Pedra 22": primeiro_nao_nulo,
        "Pedra 23": primeiro_nao_nulo,
        "Pedra 2024": primeiro_nao_nulo
    })
)

# score ordinal das pedras
mapa_score = {
    "Quartzo": 1,
    "Ágata": 2,
    "Ametista": 3,
    "Topázio": 4
}

df_consulta_pedras["score_22"] = df_consulta_pedras["Pedra 22"].map(mapa_score)
df_consulta_pedras["score_23"] = df_consulta_pedras["Pedra 23"].map(mapa_score)
df_consulta_pedras["score_24"] = df_consulta_pedras["Pedra 2024"].map(mapa_score)

# análise da trajetória
def analisar_trajetoria(row):
    s22 = row["score_22"]
    s23 = row["score_23"]
    s24 = row["score_24"]

    scores = [x for x in [s22, s23, s24] if pd.notna(x)]

    if len(scores) < 2:
        return "Informação insuficiente"

    if scores == sorted(scores) and len(set(scores)) == 1:
        return "Estável"

    if scores == sorted(scores) and len(set(scores)) > 1:
        return "Melhora consistente"

    if scores == sorted(scores, reverse=True) and len(set(scores)) > 1:
        return "Piora consistente"

    if scores[-1] > scores[0]:
        return "Oscilou, mas melhorou no final"

    if scores[-1] < scores[0]:
        return "Oscilou, mas piorou no final"

    return "Oscilou e terminou estável"

df_consulta_pedras["analise"] = df_consulta_pedras.apply(analisar_trajetoria, axis=1)

print("Dimensão da base final:")
print(df_consulta_pedras.shape)

print("\nAmostra:")
display(df_consulta_pedras.head(20))

print("\nResumo da análise:")
display(
    df_consulta_pedras["analise"]
    .value_counts(dropna=False)
    .rename_axis("analise")
    .reset_index(name="quantidade")
)
```

- manutenção do score = estabilidade
- redução do score = regressão

- **Melhora consistente**

- **Oscilou, mas melhorou no final**
- **Oscilou, mas piorou no final**
- **Oscilou e terminou estável**
- **Informação insuficiente**
