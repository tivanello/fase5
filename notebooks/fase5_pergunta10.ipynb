```python
# ============================================================
# BLOCO P10.15 — VALIDAR PEDRA 23 x PEDRA 2023 POR RA
# Objetivo:
# - consolidar por RA
# - trazer Pedra 23 e Pedra 2023 lado a lado
# - classificar a situação entre as duas colunas
# ============================================================

df_tmp = df_fase5_p10[["RA", "Pedra 23", "Pedra 2023"]].copy()

# padroniza texto
for c in ["Pedra 23", "Pedra 2023"]:
    df_tmp[c] = df_tmp[c].astype("string").str.strip()

# função para pegar o primeiro valor não nulo
def primeiro_nao_nulo(serie):
    serie = serie.dropna()
    return serie.iloc[0] if len(serie) > 0 else pd.NA

# consolida por RA
df_validacao_2023 = (
    df_tmp
    .groupby("RA", as_index=False)
    .agg({
        "Pedra 23": primeiro_nao_nulo,
        "Pedra 2023": primeiro_nao_nulo
    })
)

# coluna de validação
def validar_pedra_2023(row):
    p23 = row["Pedra 23"]
    p2023 = row["Pedra 2023"]

    if pd.isna(p23) and pd.isna(p2023):
        return "AMBAS_VAZIAS"
    elif pd.notna(p23) and pd.notna(p2023) and p23 == p2023:
        return "IGUAL"
    elif pd.notna(p23) and pd.isna(p2023):
        return "COMPLEMENTAR_SO_PEDRA23"
    elif pd.isna(p23) and pd.notna(p2023):
        return "COMPLEMENTAR_SO_PEDRA2023"
    else:
        return "CONFLITO"

df_validacao_2023["validacao"] = df_validacao_2023.apply(validar_pedra_2023, axis=1)

print("Resumo da validação por RA:")
display(
    df_validacao_2023["validacao"]
    .value_counts(dropna=False)
    .rename_axis("validacao")
    .reset_index(name="quantidade")
)

print("\nAmostra da base consolidada por RA:")
display(df_validacao_2023.head(30))
```

```python
# ============================================================
# BLOCO P10.15 — VALIDAR PEDRA 23 x PEDRA 2023 POR RA
# Objetivo:
# - consolidar por RA
# - trazer Pedra 23 e Pedra 2023 lado a lado
# - classificar a situação entre as duas colunas
# ============================================================

df_tmp = df_fase5_p10[["RA", "Pedra 23", "Pedra 2023"]].copy()

# padroniza texto
for c in ["Pedra 23", "Pedra 2023"]:
    df_tmp[c] = df_tmp[c].astype("string").str.strip()

# função para pegar o primeiro valor não nulo
def primeiro_nao_nulo(serie):
    serie = serie.dropna()
    return serie.iloc[0] if len(serie) > 0 else pd.NA

# consolida por RA
df_validacao_2023 = (
    df_tmp
    .groupby("RA", as_index=False)
    .agg({
        "Pedra 23": primeiro_nao_nulo,
        "Pedra 2023": primeiro_nao_nulo
    })
)

# coluna de validação
def validar_pedra_2023(row):
    p23 = row["Pedra 23"]
    p2023 = row["Pedra 2023"]

    if pd.isna(p23) and pd.isna(p2023):
        return "AMBAS_VAZIAS"
    elif pd.notna(p23) and pd.notna(p2023) and p23 == p2023:
        return "IGUAL"
    elif pd.notna(p23) and pd.isna(p2023):
        return "COMPLEMENTAR_SO_PEDRA23"
    elif pd.isna(p23) and pd.notna(p2023):
        return "COMPLEMENTAR_SO_PEDRA2023"
    else:
        return "CONFLITO"

df_validacao_2023["validacao"] = df_validacao_2023.apply(validar_pedra_2023, axis=1)

print("Resumo da validação por RA:")
display(
    df_validacao_2023["validacao"]
    .value_counts(dropna=False)
    .rename_axis("validacao")
    .reset_index(name="quantidade")
)
```

```python
# ============================================================
# BLOCO — TRANSFORMAR AS PEDRAS EM BASE LONGA POR ALUNO
# Objetivo:
# - reorganizar as colunas de pedra em formato longitudinal
# - manter apenas registros com pedra preenchida
# ============================================================

df_base = df_fase5_p10.copy()

colunas_pedra = {
    "Pedra 21": 1,
    "Pedra 22": 2,
    "Pedra 23": 3,
    "Pedra 2024": 4,
    "PEDRA_final": 5
}

bases = []

for col, ordem in colunas_pedra.items():
    df_temp = df_base[["RA", col]].copy()
    df_temp = df_temp.rename(columns={col: "pedra"})
    df_temp["pedra_referencia"] = col
    df_temp["ordem_referencia"] = ordem
    bases.append(df_temp)

df_pedra_long = pd.concat(bases, ignore_index=True)

df_pedra_long = df_pedra_long[["RA", "pedra_referencia", "ordem_referencia", "pedra"]].copy()

# remove linhas sem pedra
df_pedra_long = df_pedra_long[df_pedra_long["pedra"].notna()].copy()

print("Dimensão da base longa:")
print(df_pedra_long.shape)

print("\nAmostra:")
display(df_pedra_long.head(20))

print("\nNulos por coluna:")
print(df_pedra_long.isna().sum())
```

```python
# ============================================================
# BLOCO — LISTAR AS PEDRAS EXISTENTES
# Objetivo:
# - mostrar quais valores existem na coluna de pedra
# - incluir a quantidade de ocorrências de cada uma
# ============================================================

print("Pedras existentes:")
display(df_pedra_long["pedra"].value_counts(dropna=False).rename_axis("pedra").reset_index(name="quantidade"))
```

```python
# ============================================================
# BLOCO — PADRONIZAR AS PEDRAS DA BASE LONGA E CRIAR SCORE
# Objetivo:
# - corrigir grafias diferentes da mesma pedra
# - remover valores inválidos como INCLUIR
# - criar score ordinal para análise de evolução
# ============================================================

df_pedra_long = df_pedra_long.copy()

df_pedra_long["pedra"] = df_pedra_long["pedra"].astype("string").str.strip()

df_pedra_long["pedra"] = df_pedra_long["pedra"].replace({
    "Agata": "Ágata",
    "AGATA": "Ágata",
    "Topazio": "Topázio",
    "TOPAZIO": "Topázio",
    "INCLUIR": pd.NA
})

df_pedra_long = df_pedra_long[df_pedra_long["pedra"].notna()].copy()

mapa_score_pedra = {
    "Quartzo": 1,
    "Ágata": 2,
    "Ametista": 3,
    "Topázio": 4
}

df_pedra_long["score_pedra"] = df_pedra_long["pedra"].map(mapa_score_pedra)

print("Pedras existentes após padronização:")
display(
    df_pedra_long["pedra"]
    .value_counts(dropna=False)
    .rename_axis("pedra")
    .reset_index(name="quantidade")
)

print("\nAmostra com score_pedra:")
display(df_pedra_long.head(20))

print("\nNulos em score_pedra:")
print(df_pedra_long["score_pedra"].isna().sum())
```

```python
# ============================================================
# BLOCO — GERAR O SCORE FINAL DA PEDRA FINAL POR ALUNO
# Objetivo:
# - pegar a linha da PEDRA_final
# - gerar o score final de cada aluno
# ============================================================

df_score_final = (
    df_pedra_long[df_pedra_long["pedra_referencia"] == "PEDRA_final"]
    [["RA", "pedra", "score_pedra"]]
    .copy()
    .rename(columns={
        "pedra": "pedra_final",
        "score_pedra": "score_pedra_final"
    })
    .sort_values("RA")
    .reset_index(drop=True)
)

print("Dimensão da base de score final:")
print(df_score_final.shape)

print("\nAmostra:")
display(df_score_final.head(20))

print("\nDistribuição do score final:")
display(
    df_score_final["score_pedra_final"]
    .value_counts(dropna=False)
    .sort_index()
    .rename_axis("score_pedra_final")
    .reset_index(name="quantidade")
)
```

```python
# ============================================================
# BLOCO — GERAR SCORE INICIAL, FINAL E VARIAÇÃO POR ALUNO
# Objetivo:
# - identificar a primeira e a última pedra de cada aluno
# - calcular a variação final da trajetória
# ============================================================

df_traj = df_pedra_long.copy()

df_traj = df_traj.sort_values(["RA", "ordem_referencia"])

df_inicio_fim = (
    df_traj.groupby("RA")
    .agg(
        pedra_inicial=("pedra", "first"),
        score_inicial=("score_pedra", "first"),
        pedra_final=("pedra", "last"),
        score_final=("score_pedra", "last")
    )
    .reset_index()
)

df_inicio_fim["variacao_score"] = df_inicio_fim["score_final"] - df_inicio_fim["score_inicial"]

print("Amostra da trajetória inicial e final:")
display(df_inicio_fim.head(20))

print("\nDistribuição da variação do score:")
display(
    df_inicio_fim["variacao_score"]
    .value_counts(dropna=False)
    .sort_index()
    .rename_axis("variacao_score")
    .reset_index(name="quantidade")
)
```

```python
# ============================================================
# BLOCO 3 — GERAR TRAJETÓRIA INICIAL E FINAL POR ALUNO
# Objetivo:
# - Identificar primeira e última pedra válida de cada aluno
# ============================================================

df_pedra_valida = df_pedra_valida.sort_values(["RA", "ano_referencia"])

df_inicio = (
    df_pedra_valida.groupby("RA", as_index=False)
    .first()[["RA", "ano_referencia", "pedra", "score_pedra"]]
    .rename(columns={
        "ano_referencia": "ano_inicial",
        "pedra": "pedra_inicial",
        "score_pedra": "score_inicial"
    })
)

df_fim = (
    df_pedra_valida.groupby("RA", as_index=False)
    .last()[["RA", "ano_referencia", "pedra", "score_pedra"]]
    .rename(columns={
        "ano_referencia": "ano_final",
        "pedra": "pedra_final",
        "score_pedra": "score_final"
    })
)

df_evolucao_aluno = df_inicio.merge(df_fim, on="RA", how="inner")

print("Dimensão da base de evolução:", df_evolucao_aluno.shape)

print("\nAmostra:")
display(df_evolucao_aluno.head(20))
```

```python
# ============================================================
# BLOCO 4 — CLASSIFICAR EVOLUÇÃO DO ALUNO
# Objetivo:
# - Classificar cada aluno em Melhorou, Estável ou Piorou
# ============================================================

df_evolucao_aluno = df_evolucao_aluno.copy()

df_evolucao_aluno["variacao_score"] = (
    df_evolucao_aluno["score_final"] - df_evolucao_aluno["score_inicial"]
)

def classificar_evolucao(x):
    if x > 0:
        return "Melhorou"
    elif x == 0:
        return "Estável"
    elif x < 0:
        return "Piorou"
    else:
        return pd.NA

df_evolucao_aluno["evolucao"] = df_evolucao_aluno["variacao_score"].apply(classificar_evolucao)

print("Amostra:")
display(df_evolucao_aluno.head(20))
```

```python
# ============================================================
# BLOCO 5 — RESUMO DA EVOLUÇÃO DOS ALUNOS
# Objetivo:
# - Quantificar quantos alunos melhoraram, ficaram estáveis ou pioraram
# ============================================================

df_q10_evolucao_resumo = (
    df_evolucao_aluno["evolucao"]
    .value_counts(dropna=False)
    .reset_index()
)

df_q10_evolucao_resumo.columns = ["evolucao", "qtd_alunos"]
df_q10_evolucao_resumo["percentual"] = (
    df_q10_evolucao_resumo["qtd_alunos"] / df_q10_evolucao_resumo["qtd_alunos"].sum() * 100
).round(2)

display(df_q10_evolucao_resumo)
```

```python
# ============================================================
# BLOCO — CLASSIFICAR CONSISTÊNCIA DA EVOLUÇÃO
# Objetivo:
# - Avaliar a trajetória ano a ano de cada aluno
# - Identificar se houve melhora consistente, estabilidade ou regressão
# ============================================================

import pandas as pd
import numpy as np

df_base = df_pedra_long.copy()

# Padronizar pedras
df_base["pedra"] = df_base["pedra"].replace({
    "Agata": "Ágata",
    "Topazio": "Topázio",
    "INCLUIR": pd.NA
})

# Criar score
mapa_score_pedra = {
    "Quartzo": 1,
    "Ágata": 2,
    "Ametista": 3,
    "Topázio": 4
}

df_base["score_pedra"] = df_base["pedra"].map(mapa_score_pedra)

# Manter apenas registros válidos
df_base = df_base.dropna(subset=["score_pedra"]).copy()

# Ordenar
df_base = df_base.sort_values(["RA", "ano_referencia"])

def classificar_trajetoria(grupo):
    scores = grupo["score_pedra"].tolist()

    if len(scores) <= 1:
        return "Sem histórico suficiente"

    variacoes = np.diff(scores)

    houve_queda = (variacoes < 0).any()
    houve_subida = (variacoes > 0).any()

    if not houve_queda and houve_subida:
        return "Melhora consistente"
    elif not houve_queda and not houve_subida:
        return "Estável"
    elif houve_queda:
        return "Com regressão"
    else:
        return "Não classificado"

df_consistencia = (
    df_base.groupby("RA")
    .apply(classificar_trajetoria)
    .reset_index(name="classificacao_trajetoria")
)

print("Amostra:")
display(df_consistencia.head(20))
```

```python
# ============================================================
# BLOCO — RESUMO DA CONSISTÊNCIA DA MELHORA
# Objetivo:
# - Quantificar quantos alunos tiveram melhora consistente,
#   estabilidade, regressão ou histórico insuficiente
# ============================================================

df_consistencia_resumo = (
    df_consistencia["classificacao_trajetoria"]
    .value_counts(dropna=False)
    .reset_index()
)

df_consistencia_resumo.columns = ["classificacao_trajetoria", "qtd_alunos"]
df_consistencia_resumo["percentual"] = (
    df_consistencia_resumo["qtd_alunos"] / df_consistencia_resumo["qtd_alunos"].sum() * 100
).round(2)

display(df_consistencia_resumo)
```

```python
# ============================================================
# BLOCO — MONTAR TRAJETÓRIA DO ALUNO
# Objetivo:
# - Mostrar a sequência de pedras e scores de cada aluno
# ============================================================

df_trajetoria_aluno = (
    df_base.groupby("RA")
    .agg(
        anos=("ano_referencia", lambda x: list(x)),
        pedras=("pedra", lambda x: list(x)),
        scores=("score_pedra", lambda x: list(x))
    )
    .reset_index()
)

df_trajetoria_aluno = df_trajetoria_aluno.merge(df_consistencia, on="RA", how="left")

display(df_trajetoria_aluno.head(20))
```

## Justificativa para uso da variável **Instituição de ensino** com 3 categorias

Para a construção do novo modelo preditivo, foi mantida apenas a variável original **Instituição de ensino**, por ser a coluna efetivamente disponível e diretamente interpretável no contexto educacional do aluno.

Após a análise dos valores existentes, verificou-se que a coluna apresentava categorias com o mesmo significado prático, mas registradas com textos diferentes, além de alguns grupos com baixa frequência. Esse cenário poderia gerar fragmentação desnecessária da informação e reduzir a estabilidade da variável no modelo.

Por esse motivo, optou-se por padronizar a coluna em apenas **3 categorias analíticas**:

- **Pública**
- **Privada**
- **Outras**

### Regras adotadas

**Pública**
- Pública
- Escola Pública

**Privada**
- Privada
- Privada - Programa de Apadrinhamento
- Privada - Programa de apadrinhamento
- Privada *Parcerias com Bolsa 100%
- Privada - Pagamento por *Empresa Parceira

**Outras**
- Rede Decisão
- Escola JP II
- Concluiu o 3º EM
- Bolsista Universitário *Formado (a)
- Nenhuma das opções acima
- valor nulo

### Fundamentação da decisão

A adoção dessas 3 categorias foi feita pelos seguintes motivos:

- eliminar duplicidades semânticas, como **Pública** e **Escola Pública**
- unificar diferentes formas de vínculo com ensino privado em uma única categoria mais robusta
- reduzir a pulverização causada por categorias com baixa ocorrência
- evitar excesso de cardinalidade em uma variável categórica
- melhorar a consistência da informação para uso em modelagem preditiva
- preservar a leitura analítica da origem educacional do aluno sem criar detalhamento excessivo

Além disso, como havia apenas **1 valor nulo** em toda a base, esse registro foi incorporado à categoria **Outras**, por ter impacto estatístico irrelevante e não justificar a criação de uma categoria específica apenas para ausência de preenchimento.

Dessa forma, a variável **Instituição de ensino** foi mantida no modelo em versão padronizada, com estrutura mais simples, mais estável e mais adequada para representar o contexto escolar do aluno.
