```python
# ============================================================
# BLOCO 4.2 — VALIDAR PEDRA 23 x PEDRA 2023 POR RA
# Objetivo:
# - consolidar por RA
# - trazer Pedra 23 e Pedra 2023 lado a lado
# - classificar a situação entre as duas colunas
# ============================================================

df_tmp = df_fase5_p10[["RA", "Pedra 23", "Pedra 2023"]].copy()

# padroniza texto
for c in ["Pedra 23", "Pedra 2023"]:
    df_tmp[c] = df_tmp[c].astype("string").str.strip()

# função para pegar o primeiro valor não nulo
def primeiro_nao_nulo(serie):
    serie = serie.dropna()
    return serie.iloc[0] if len(serie) > 0 else pd.NA

# consolida por RA
df_validacao_2023 = (
    df_tmp
    .groupby("RA", as_index=False)
    .agg({
        "Pedra 23": primeiro_nao_nulo,
        "Pedra 2023": primeiro_nao_nulo
    })
)

# coluna de validação
def validar_pedra_2023(row):
    p23 = row["Pedra 23"]
    p2023 = row["Pedra 2023"]

    if pd.isna(p23) and pd.isna(p2023):
        return "AMBAS_VAZIAS"
    elif pd.notna(p23) and pd.notna(p2023) and p23 == p2023:
        return "IGUAL"
    elif pd.notna(p23) and pd.isna(p2023):
        return "COMPLEMENTAR_SO_PEDRA23"
    elif pd.isna(p23) and pd.notna(p2023):
        return "COMPLEMENTAR_SO_PEDRA2023"
    else:
        return "CONFLITO"

df_validacao_2023["validacao"] = df_validacao_2023.apply(validar_pedra_2023, axis=1)

print("Resumo da validação por RA:")
display(
    df_validacao_2023["validacao"]
    .value_counts(dropna=False)
    .rename_axis("validacao")
    .reset_index(name="quantidade")
)

print("\nAmostra da base consolidada por RA:")
display(df_validacao_2023.head(30))
```

```python
# ============================================================
# BLOCO 5.1 — BASE LONGA DE PEDRAS (2022 A 2024) COM PADRONIZAÇÃO E SCORE
# Objetivo:
# - considerar apenas 2022, 2023 e 2024
# - padronizar nomes das pedras
# - remover valores inválidos
# - criar score ordinal para análise de evolução
# ============================================================

df_base = df_fase5_p10.copy()

colunas_pedra = {
    "Pedra 22": 1,
    "Pedra 23": 2,
    "Pedra 24": 3
}

bases = []

for col, ordem in colunas_pedra.items():
    df_temp = df_base[["RA", col]].copy()
    df_temp = df_temp.rename(columns={col: "pedra"})
    df_temp["pedra_referencia"] = col
    df_temp["ordem_referencia"] = ordem
    bases.append(df_temp)

df_pedra_long = pd.concat(bases, ignore_index=True)

df_pedra_long = df_pedra_long[["RA", "pedra_referencia", "ordem_referencia", "pedra"]].copy()

df_pedra_long["pedra"] = df_pedra_long["pedra"].astype("string").str.strip()

df_pedra_long["pedra"] = df_pedra_long["pedra"].replace({
    "Agata": "Ágata",
    "AGATA": "Ágata",
    "Topazio": "Topázio",
    "TOPAZIO": "Topázio",
    "INCLUIR": pd.NA
})

df_pedra_long = df_pedra_long[df_pedra_long["pedra"].notna()].copy()

mapa_score_pedra = {
    "Quartzo": 1,
    "Ágata": 2,
    "Ametista": 3,
    "Topázio": 4
}

df_pedra_long["score_pedra"] = df_pedra_long["pedra"].map(mapa_score_pedra)

print("Dimensão da base longa:")
print(df_pedra_long.shape)

print("\nAmostra:")
display(df_pedra_long.tail(20))

print("\nDistribuição das pedras:")
display(
    df_pedra_long["pedra"]
    .value_counts(dropna=False)
    .rename_axis("pedra")
    .reset_index(name="quantidade")
)

print("\nNulos em score_pedra:")
print(df_pedra_long["score_pedra"].isna().sum())
```

```python
# ============================================================
# BLOCO 5.2 — CONSULTA DAS PEDRAS POR RA COM ANÁLISE DOS SCORES
# Objetivo:
# - montar uma visão por aluno com as pedras de 2022, 2023 e 2024
# - converter as pedras em scores ordinais
# - classificar a trajetória do aluno no período
# ============================================================

df_consulta_pedras = df_fase5_p10.copy()

colunas = ["RA", "Pedra 22", "Pedra 23", "Pedra 24"]
df_consulta_pedras = df_consulta_pedras[colunas].copy()

# padroniza texto
for c in ["Pedra 22", "Pedra 23", "Pedra 24"]:
    df_consulta_pedras[c] = df_consulta_pedras[c].astype("string").str.strip()

for c in ["Pedra 22", "Pedra 23", "Pedra 24"]:
    df_consulta_pedras[c] = df_consulta_pedras[c].replace(mapa_padrao)

# função para pegar o primeiro valor não nulo por RA
def primeiro_nao_nulo(s):
    s = s.dropna()
    return s.iloc[0] if len(s) > 0 else pd.NA

# consolida por RA
df_consulta_pedras = (
    df_consulta_pedras
    .groupby("RA", as_index=False)
    .agg({
        "Pedra 22": primeiro_nao_nulo,
        "Pedra 23": primeiro_nao_nulo,
        "Pedra 24": primeiro_nao_nulo
    })
)

# score ordinal das pedras
mapa_score = {
    "Quartzo": 1,
    "Ágata": 2,
    "Ametista": 3,
    "Topázio": 4
}

df_consulta_pedras["score_22"] = df_consulta_pedras["Pedra 22"].map(mapa_score)
df_consulta_pedras["score_23"] = df_consulta_pedras["Pedra 23"].map(mapa_score)
df_consulta_pedras["score_24"] = df_consulta_pedras["Pedra 24"].map(mapa_score)

# análise da trajetória
def analisar_trajetoria(row):
    s22 = row["score_22"]
    s23 = row["score_23"]
    s24 = row["score_24"]

    scores = [x for x in [s22, s23, s24] if pd.notna(x)]

    if len(scores) < 2:
        return "Informação insuficiente"

    if scores == sorted(scores) and len(set(scores)) == 1:
        return "Estável"

    if scores == sorted(scores) and len(set(scores)) > 1:
        return "Melhora consistente"

    if scores == sorted(scores, reverse=True) and len(set(scores)) > 1:
        return "Piora consistente"

    if scores[-1] > scores[0]:
        return "Oscilou, mas melhorou no final"

    if scores[-1] < scores[0]:
        return "Oscilou, mas piorou no final"

    return "Oscilou e terminou estável"

df_consulta_pedras["analise"] = df_consulta_pedras.apply(analisar_trajetoria, axis=1)

print("Dimensão da base final:")
print(df_consulta_pedras.shape)

print("\nAmostra:")
display(df_consulta_pedras.head(20))

print("\nResumo da análise:")
display(
    df_consulta_pedras["analise"]
    .value_counts(dropna=False)
    .rename_axis("analise")
    .reset_index(name="quantidade")
)
```

```python
# ============================================================
# BLOCO 6 — BASE FINAL DA P10 COM ELEGIBILIDADE
# Objetivo:
# - contar quantos pontos válidos cada aluno possui entre 2022 e 2024
# - definir se o aluno é elegível para análise longitudinal
# - manter a base final pronta para exportação e uso no Power BI
# ============================================================

df_p10_base_final = df_consulta_pedras.copy()

# quantidade de pontos válidos por aluno
df_p10_base_final["qtd_pontos_validos"] = (
    df_p10_base_final[["score_22", "score_23", "score_24"]]
    .notna()
    .sum(axis=1)
)

# regra de elegibilidade
df_p10_base_final["elegivel_analise"] = np.where(
    df_p10_base_final["qtd_pontos_validos"] >= 2,
    "Sim",
    "Não"
)

# análise resumida
def resumir_analise(valor):
    if valor in ["Melhora consistente", "Oscilou, mas melhorou no final"]:
        return "Melhorou"
    elif valor in ["Estável", "Oscilou e terminou estável"]:
        return "Estável"
    elif valor in ["Piora consistente", "Oscilou, mas piorou no final"]:
        return "Piorou"
    else:
        return "Informação insuficiente"

df_p10_base_final["analise_resumida"] = df_p10_base_final["analise"].apply(resumir_analise)

# organiza colunas finais
df_p10_base_final = df_p10_base_final[
    [
        "RA",
        "Pedra 22",
        "Pedra 23",
        "Pedra 24",
        "score_22",
        "score_23",
        "score_24",
        "qtd_pontos_validos",
        "elegivel_analise",
        "analise",
        "analise_resumida"
    ]
].copy()

print("Dimensão da base final da P10:")
print(df_p10_base_final.shape)

print("\nAmostra:")
display(df_p10_base_final.head(20))

print("\nResumo da elegibilidade:")
display(
    df_p10_base_final["elegivel_analise"]
    .value_counts(dropna=False)
    .rename_axis("elegivel_analise")
    .reset_index(name="quantidade")
)

print("\nResumo da análise resumida:")
display(
    df_p10_base_final["analise_resumida"]
    .value_counts(dropna=False)
    .rename_axis("analise_resumida")
    .reset_index(name="quantidade")
)
```

```python
# ============================================================
# BLOCO 6.1 — GERAR RESUMO EXECUTIVO DA P10
# Objetivo:
# - consolidar a análise da P10 em contagem e percentual
# - separar base total e base elegível para análise longitudinal
# - gerar tabelas prontas para uso no Power BI
# ============================================================

# ------------------------------
# 1. Resumo da base total
# ------------------------------
df_p10_resumo_total = (
    df_p10_base_final["analise_resumida"]
    .value_counts(dropna=False)
    .rename_axis("analise_resumida")
    .reset_index(name="quantidade")
)

total_alunos = df_p10_resumo_total["quantidade"].sum()

df_p10_resumo_total["percentual"] = (
    df_p10_resumo_total["quantidade"] / total_alunos * 100
).round(2)

df_p10_resumo_total["base"] = "Base total 2022-2024"

df_p10_resumo_total = df_p10_resumo_total[
    ["base", "analise_resumida", "quantidade", "percentual"]
].copy()

print("Resumo executivo — base total:")
display(df_p10_resumo_total)

# ------------------------------
# 2. Resumo apenas da base elegível
# ------------------------------
df_p10_elegiveis = df_p10_base_final[
    df_p10_base_final["elegivel_analise"] == "Sim"
].copy()

df_p10_resumo_elegiveis = (
    df_p10_elegiveis["analise_resumida"]
    .value_counts(dropna=False)
    .rename_axis("analise_resumida")
    .reset_index(name="quantidade")
)

total_elegiveis = df_p10_resumo_elegiveis["quantidade"].sum()

df_p10_resumo_elegiveis["percentual"] = (
    df_p10_resumo_elegiveis["quantidade"] / total_elegiveis * 100
).round(2)

df_p10_resumo_elegiveis["base"] = "Base elegível"

df_p10_resumo_elegiveis = df_p10_resumo_elegiveis[
    ["base", "analise_resumida", "quantidade", "percentual"]
].copy()

print("\nResumo executivo — base elegível:")
display(df_p10_resumo_elegiveis)

# ------------------------------
# 3. Resumo consolidado
# ------------------------------
df_p10_resumo_analise = pd.concat(
    [df_p10_resumo_total, df_p10_resumo_elegiveis],
    ignore_index=True
)

print("\nResumo executivo consolidado:")
display(df_p10_resumo_analise)

# ------------------------------
# 4. Resumo da elegibilidade
# ------------------------------
df_p10_resumo_elegibilidade = (
    df_p10_base_final["elegivel_analise"]
    .value_counts(dropna=False)
    .rename_axis("elegivel_analise")
    .reset_index(name="quantidade")
)

total_base = df_p10_resumo_elegibilidade["quantidade"].sum()

df_p10_resumo_elegibilidade["percentual"] = (
    df_p10_resumo_elegibilidade["quantidade"] / total_base * 100
).round(2)

print("\nResumo da elegibilidade:")
display(df_p10_resumo_elegibilidade)
```

A Pergunta 10 busca verificar se os indicadores educacionais evidenciam melhora consistente ao longo do ciclo, nas diferentes fases representadas pelas pedras **Quartzo**, **Ágata**, **Ametista** e **Topázio**, confirmando o impacto real do programa.

Para responder a essa questão, foi construída uma análise longitudinal por aluno no período oficial do Datathon, considerando exclusivamente os anos de **2022, 2023 e 2024**, por meio das colunas **Pedra 22**, **Pedra 23** e **Pedra 24**.

As pedras foram convertidas em uma escala ordinal para permitir comparação objetiva entre os anos:

- **Quartzo = 1**
- **Ágata = 2**
- **Ametista = 3**
- **Topázio = 4**

Os registros com valor nulo e os casos classificados como **INCLUIR** foram desconsiderados na análise das trajetórias. Essa exclusão foi necessária porque, nessas situações, não havia informação válida suficiente para identificar com segurança a pedra correspondente e, portanto, não era possível atribuir um score confiável para comparação longitudinal.

Com base nessa escala, cada aluno foi classificado segundo sua trajetória no período, distinguindo situações de:

- **Melhora consistente**
- **Estabilidade**
- **Piora consistente**
- **Oscilação com melhora final**
- **Oscilação com piora final**
- **Oscilação com término estável**
- **Informação insuficiente**

Além da análise detalhada, foi gerado um resumo executivo com a consolidação dos resultados em categorias sintéticas, permitindo identificar o saldo geral das trajetórias dos alunos no ciclo 2022–2024.

Essa etapa é fundamental para responder à Pergunta 10, pois desloca a análise de uma fotografia isolada da pedra final para uma leitura de trajetória, tornando possível avaliar se houve avanço real, manutenção de desempenho ou regressão ao longo do período analisado.

O resultado dessa consolidação servirá como base para os visuais e indicadores do Power BI, especialmente para:

- distribuição dos alunos por tipo de trajetória
- percentual de melhora, estabilidade e piora
- leitura executiva sobre a efetividade do programa no ciclo analisado

Como limitação metodológica, destaca-se que alunos com quantidade insuficiente de registros válidos no período são classificados como **informação insuficiente**, não sendo interpretados como evidência de melhora, estabilidade ou piora.

### **Recorte temporal e elegibilidade da análise**

A análise principal da Pergunta 10 considera exclusivamente o período de **2022 a 2024**, em conformidade com o recorte oficial do Datathon. Para isso, foram utilizadas as colunas **Pedra 22**, **Pedra 23** e **Pedra 24**.

Dentro desse período, a base foi separada em dois conjuntos:

- **base total do recorte 2022–2024**: contempla todos os alunos com registros no período
- **amostra elegível para análise de trajetória**: contempla apenas os alunos com informações suficientes para comparação longitudinal entre os anos

Essa distinção é importante porque nem todos os alunos possuem registros válidos em quantidade suficiente para permitir classificação segura de melhora, estabilidade ou regressão. Nesses casos, o aluno permanece na base do período, mas é classificado como **informação insuficiente** na análise executiva.
